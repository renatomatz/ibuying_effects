```python
df =pd.read_html("https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations")[0]
```

```python
df = df.iloc[11:, [0, 3]].dropna().reset_index(drop=True)
df.columns = ["state", "abbreviation"]
df
```
